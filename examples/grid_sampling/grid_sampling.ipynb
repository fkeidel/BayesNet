# Markov Random Field Grid Sampling 

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

file_path = "C:/BayesNet/examples/grid_sampling/plots/"

def plot_image(file_name, rows, columns, index):
    plt.subplot(rows, columns, index)
    img = np.loadtxt(file_name, delimiter=',')
    plt.imshow(img, vmin=0, vmax=1)
    plt.xticks(np.arange(0,4,1))
    plt.yticks(np.arange(0,4,1))

def plot_image_singleton_factors(file_path):
    plt.figure()
    plot_image(file_path + "SingletonFactors.csv", 1,1,1)
    plt.colorbar()
    plt.title("Singleton Factors")
    plt.show()

def plot_image_exact_marginals(file_path):
    plt.figure()
    plot_image(file_path + "Image_Exact_0.csv", 1,1,1)
    plt.colorbar()
    plt.title("Exact marginals")
    plt.show()

def plot_images_est_marginals(file_path, method):
    plt.figure(figsize=(10,5))
    for i in range(0,3):
        plot_image(file_path + "Image_" + method + "_" + str(i) + ".csv", 1,3,i+1)
    cax = plt.axes([0.95, 0.25, 0.016, 0.5])
    plt.colorbar(cax=cax)
    plt.show()

def plot_chart_est_marginals(file_path, method):
    var = 2
    plt.figure()
    for i in range(0,3):
        chart = np.loadtxt(file_path + "Chart_Est_Marginals_Var" + str(var) + "_" + method + "_" + str(i) + ".csv", skiprows=1, delimiter=',')
        plt.plot(chart[:,1]) # plot "true" values (second column)
    plt.title(f"Est. marginals for variable {var}=true")
    plt.xlabel("cycle")
    plt.show()

def plot_likelihoods(file_path, method):
    likelihoods = np.loadtxt(file_path + "Likelihoods_" + method + ".csv", skiprows=1, delimiter=',')
    plt.figure()
    plt.plot(likelihoods)
    plt.title("Likelihood")
    plt.xlabel("cycle")
    plt.ylabel("log")
    plt.show()

def plot_mcmc_data(file_path, method):
    plot_images_est_marginals(file_path, method)
    plot_chart_est_marginals(file_path, method)
    plot_likelihoods(file_path, method)
```

```python
plot_image_singleton_factors(file_path)
```
