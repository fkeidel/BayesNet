# Markov Random Field Grid Sampling 

```python
import matplotlib.pyplot as plt
import numpy as np

file_path = "C:/BayesNet/examples/grid_sampling/plots/"

def plot_image(file_name, rows, columns, index):
    plt.subplot(rows, columns, index)
    img = np.loadtxt(file_name, delimiter=',')
    plt.imshow(img, vmin=0, vmax=1)
    plt.xticks(np.arange(0,4,1))
    plt.yticks(np.arange(0,4,1))

def plot_charts(file_path, var, method):
    for i in range(0,3):
        chart = np.loadtxt(file_path + "Chart_Est_Marginals_Var" + str(var) + "_" + method + "_" + str(i) + ".csv", skiprows=1, delimiter=',')
        plt.plot(chart[:,1]) # plot "true" values (second column)

plt.figure()
plot_image(file_path + "Image_Exact_0.csv", 1,1,1)
plt.colorbar()
plt.title("Exact marginals")
plt.show()

method = "MHSwendsenWang2"

plt.figure(figsize=(10,5))
for i in range(0,3):
    plot_image(file_path + "Image_" + method + "_" + str(i) + ".csv", 1,3,i+1)
cax = plt.axes([0.95, 0.25, 0.016, 0.5])
plt.colorbar(cax=cax)
plt.show()

plt.figure()
plot_charts(file_path, 2, method)
plt.title("Approx. marginals ('true' value)")
plt.xlabel("cycle")
plt.show()

likelihoods = np.loadtxt(file_path + "Likelihoods_" + method + ".csv", skiprows=1, delimiter=',')
plt.figure()
plt.plot(likelihoods)
plt.title("Likelihood")
plt.xlabel("cycle")
plt.ylabel("log")
plt.show()

```
